```python
import numpy as np
x=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,0,0,1])
```

```python
weights=np.array([0.0, 0.0])
bias=0
```

```python
for i in range(len(x)):
 x1,x2=x[i]
 target=y[i]
 z=weights[0]*x1+weights[1]*x2+bias
 output = 1 if z >= 0 else 0
 if output != target:
   if target == 1:
      weights[0] += x1
      weights[1] += x2
      bias += 1
   else:
      weights[0] -= x1
      weights[1] -= x2
      bias -= 1

```

```python
print("Weights after one iteration:", weights)
print("Bias after one iteration:", bias)
```

```python
epochs = 0
while True:
  error_count = 0
  for i in range(len(x)):
    x1,x2 =x[i]
    target=y[i]
    z=weights[0]*x1+weights[1]*x2+bias
    output = 1 if z >= 0 else 0
    if output != target:
      if target == 1:
         weights[0] += x1
         weights[1] += x2
         bias += 1
      else:
         weights[0] -= x1
         weights[1] -= x2
         bias -= 1
      error_count += 1
  epochs += 1
  if error_count == 0:
    break
```

```python
print("Training complete in",epochs,"epochs")
print("Final Weights:",weights)
print("Final Bias:", bias)

print("\nTesting on all inputs:")
for i in range(len(x)):
    x1,x2= x[i]
    z=weights[0]* x1 +weights[1]*x2 + bias
    output = 1 if z >= 0 else 0
    print(f"Input: {x[i]} -> Output: {output}")
```
